```python
df=yf.download('QQQ',start='2020-01-01')
df
```

```python
#The DataFrame object is passed to the QuantFig constructor.
qf=cf.QuantFig(df[['Open','High','Low','Close','Volume']],
              title='QQQ since 2020',
              legend='right',
              name='QQQ') # when we pass has to include the open high low 
```

```python
qf.iplot()
```

```python
# since qf is the instance of class or object
qf.add_bollinger_bands(periods=15,boll_std=2)
```

```python
qf.iplot()
```

```python
qf.add_rsi(periods=4,
          showbands=True)
```

```python
qf.iplot()
```

```python
df['Volume'].iplot()
```

```python
qf2=cf.QuantFig(df[['Open','High','Low','Close','Volume']],
              title='QQQ since 2020',
              legend='right',
              name='QQQ volume study')
```

```python
qf2.add_volume(up_color='green',down_color='red')
```

```python
qf2.iplot()
```
