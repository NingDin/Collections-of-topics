````python
import scipy.optimize as spo
````

````python
# example binominal function
# find minimize value for y
def f(x):
    y=(x**2)-(12*x)+20
    return y
````

````python
# start with a starting guess 
x_start=2.0
# start the optimazing
# note the result below from scipy will be an object
result=spo.minimize(f,x_start) # first is the function, next arguement will be x_start
# print result
if result.success:
    print("Sucess!")
    print(f"x={result.x} y={result.fun}") # fun means function
else:
    print("sorry could not find the minimum")
````

````python
def f(xy):
    x=xy[0]
    y=xy[1]
    area=x*y
    return -area
````

 Here's a real-world example of how you can use the `f(xy)` function to calculate the area of a rectangle:

```python
# Example 1: Calculate the area of a rectangle
xy = (5, 3)  # Coordinates (x, y) of the rectangle
area = f(xy)
print("The area of the rectangle is:", area)
```

In this example, we have a rectangle with sides of length 5 units and 3 units. The coordinates `(5, 3)` represent the length and width of the rectangle. By calling the `f(xy)` function with the `xy` tuple as input, we calculate the area of the rectangle. The result is then printed, which will display the area as `15`.

You can replace the `(5, 3)` with different coordinate values to calculate the area of rectangles with different dimensions.

Note: Make sure to define the `f(xy)` function before running the example.

````python
xy_start=[50,50] # 2x,y should add to 100  # since fence is 100 unit and we look to maximze the area
# add constrian it is always show as tuple of dict
cons=({'type':'eq','fun':lambda xy:(2*xy[0]+xy[1]-100)}) # we do -100 cuz build in function wanna cons value to zero
result=spo.minimize(f,xy_start,options={'disp':True},constraints=cons)
if result.success:
    print("Sucess!")
    xy=result.x # note this x is the optimizing variable in our case it two, it is always x in scip result attribute
    x=xy[0]
    y=xy[1]
    print(f"x={x} y={y}") # fun means function
else:
    print("sorry could not find the minimum")
````
